```python
su_fatura = pd.read_excel("su_fatura_destegi_2023_aylik.xlsx")
```

```python
su_fatura.rename(columns={"m3":"faturasi_odenen_su_miktari_(m3)"},inplace=True)
```

```python
su_fatura['yil-ay'] = su_fatura['yil'].astype(str) + '-' + su_fatura['ay'].astype(str).str.zfill(2)
```

```python
su_fatura['yil-ay'] = pd.to_datetime(su_fatura['yil-ay'])
```

```python
su_fatura['yil-ay'] = su_fatura['yil-ay'].dt.to_period('M')
```

```python
su_fatura.drop(columns={"yil","ay"},inplace=True)
```

```python
su_fatura.columns
```

```python
su_fatura = su_fatura[["yil-ay","desteklenen_abone_sayisi","faturasi_odenen_su_miktari_(m3)","destek_miktari_(TL)"]]
```

```python
su_fatura.head(50)
```

```python
su_fatura.to_csv("temiz-su_fatura_destegi_2023_aylik.csv" ,header=True ,index=False)
```
